```python
# References: https://wikitech.wikimedia.org/wiki/Event_Platform/EventStreams
import json
from sseclient import SSEClient as EventSource
from kafka import KafkaConsumer

# Create consumer
consumer = KafkaConsumer(bootstrap_servers='kafka-server:9092', #Kafka server
                        auto_offset_reset='earliest', #Consume messages from the beggining
                        value_deserializer=lambda m: json.loads(m.decode('utf-8'))) #json deserializer

consumer.subscribe(['wiki-changes']) #kafka topic

for message in consumer:
    print("**************************************")
    print("**************************************")
    print("**************************************")
    print(message)
```
